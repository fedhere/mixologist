```python
import sqlite3
#http://sebastianraschka.com/Articles/2014_sqlite_in_python_tutorial.html#creating-a-new-sqlite-database

sqlite_file = 'mixologist_db.sqlite'    # name of the sqlite database file
table_name = 'drinks_1'  # name of the table to be created
name_field = 'drinkname' # name of the column
name_type = 'STRING'  # column data type
recipe_field = 'recipe' # name of the column
recipe_type = 'STRING'  # column data type
```

```python
conn = sqlite3.connect(sqlite_file)
c = conn.cursor()

# Creating a new SQLite table with 1 column
c.execute('CREATE TABLE {tn} ({nf} {ft})'\
        .format(tn=table_name, nf=name_field, ft=name_type))

```

```python
c.execute("ALTER TABLE {tn} ADD COLUMN '{cn}' {ct} "\
        .format(tn=table_name, cn=recipe_field, ct=recipe_type))
```

```python
for i,dt in enumerate(dts):
    c.execute('INSERT OR IGNORE INTO {tn} ({cn1}, {cn2}) VALUES ("{v1}", "{v2}")'.\
            format(tn=table_name, cn1=name_field, cn2=recipe_field, v1=dt.contents[0], v2=dds[i].contents[0]))
```
